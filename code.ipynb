## Import Required Libraries

Import necessary libraries for data manipulation and file handling.

```python
import pandas as pd
import numpy as np
import os
import glob
from datetime import datetime

print("Libraries imported successfully!")
print(f"Pandas version: {pd.__version__}")
print(f"NumPy version: {np.__version__}")
```

## Locate CSV Files in Data Folder

Find all CSV files in the data directory.

```python
# Define the data folder path
data_folder = 'data'

# List all CSV files in the data folder
csv_files = glob.glob(os.path.join(data_folder, '*.csv'))

print(f"Found {len(csv_files)} CSV files in the data folder:\n")
for file in sorted(csv_files):
    file_name = os.path.basename(file)
    file_size = os.path.getsize(file) / (1024 * 1024)  # Size in MB
    print(f"  - {file_name:20s} ({file_size:.2f} MB)")
```

## Load All CSV Files

Load each CSV file into a pandas DataFrame. All files use semicolon (`;`) as delimiter.

```python
# Define CSV file names
csv_file_names = ['account.csv', 'card.csv', 'client.csv', 'disp.csv', 
                  'district.csv', 'loan.csv', 'order.csv', 'trans.csv']

# Load all CSV files into a dictionary
datasets = {}
load_errors = []

for file_name in csv_file_names:
    file_path = os.path.join(data_folder, file_name)
    try:
        # Load with semicolon delimiter
        df = pd.read_csv(file_path, sep=';', low_memory=False)
        datasets[file_name.replace('.csv', '')] = df
        print(f"Loaded {file_name:20s} - Shape: {df.shape}")
    except Exception as e:
        load_errors.append(f"{file_name}: {str(e)}")
        print(f"Error loading {file_name}: {str(e)}")

print(f"\n{'='*60}")
print(f"Successfully loaded: {len(datasets)}/{len(csv_file_names)} files")
if load_errors:
    print(f"Errors: {len(load_errors)}")
    for error in load_errors:
        print(f"  - {error}")
```

## Date Column Cleaning

Convert integer date formats (YYMMDD) to proper datetime objects for analysis.

```python
print("=" * 70)
print("DATE CLEANING - Before")
print("=" * 70)

# Show current date columns and their formats
date_info = {
    'account': {'column': 'date', 'sample': datasets['account']['date'].head(3).tolist()},
    'loan': {'column': 'date', 'sample': datasets['loan']['date'].head(3).tolist()},
    'trans': {'column': 'date', 'sample': datasets['trans']['date'].head(3).tolist()},
    'card': {'column': 'issued', 'sample': datasets['card']['issued'].head(3).tolist()}
}

for dataset_name, info in date_info.items():
    print(f"\n{dataset_name.upper()}.{info['column']}:")
    print(f"  Type: {datasets[dataset_name][info['column']].dtype}")
    print(f"  Sample values: {info['sample']}")

print("\n" + "=" * 70)
print("Converting integer dates (YYMMDD format) to datetime...")
print("=" * 70)
```

```python
def convert_int_to_date(date_int):
    """
    Convert integer date format (YYMMDD) to date only (no time).
    Example: 930101 -> 1993-01-01
    """
    date_str = str(int(date_int))
    year = int(date_str[:2])
    month = int(date_str[2:4])
    day = int(date_str[4:6])
    
    # Assume years 00-30 are 2000s, 31-99 are 1900s
    if year <= 30:
        year += 2000
    else:
        year += 1900
    
    return pd.to_datetime(f"{year}-{month:02d}-{day:02d}").date()

# Convert integer date columns
print("\nConverting ACCOUNT.date...")
datasets['account']['date'] = datasets['account']['date'].apply(convert_int_to_date)
print(f"Converted {len(datasets['account'])} rows")

print("\nConverting LOAN.date...")
datasets['loan']['date'] = datasets['loan']['date'].apply(convert_int_to_date)
print(f"Converted {len(datasets['loan'])} rows")

print("\nConverting TRANS.date...")
datasets['trans']['date'] = datasets['trans']['date'].apply(convert_int_to_date)
print(f"Converted {len(datasets['trans'])} rows")

print("\nConverting CARD.issued...")
# Card issued is already in string format like "931107 00:00:00"
# Extract YYMMDD part and convert
datasets['card']['issued'] = datasets['card']['issued'].str[:6].astype(int).apply(convert_int_to_date)
print(f"Converted {len(datasets['card'])} rows")

print("\n" + "=" * 70)
print("Date conversion completed successfully!")
print("=" * 70)
```

```python
print("\n" + "=" * 70)
print("DATE CLEANING - After")
print("=" * 70)

# Show updated date columns and their formats
for dataset_name, info in date_info.items():
    col = info['column']
    print(f"\n{dataset_name.upper()}.{col}:")
    print(f"  Type: {datasets[dataset_name][col].dtype}")
    print(f"  Sample values: {datasets[dataset_name][col].head(3).tolist()}")
    print(f"  Date range: {datasets[dataset_name][col].min()} to {datasets[dataset_name][col].max()}")

print("\n" + "=" * 70)
print("Summary: All date columns converted to date format (no time)")
print("=" * 70)
```

**Date Cleaning Summary:**
- Converted integer date format (YYMMDD) to proper date objects (no time component)
- Applied to: account.date, loan.date, trans.date, card.issued
- Year interpretation: 00-30 → 2000s, 31-99 → 1900s
- All dates now normalized for time-series analysis

## Client Birth Number Processing

Extract birthday and age from the birth_number field in the client dataset.

```python
print("=" * 70)
print("CLIENT BIRTH NUMBER - Before Processing")
print("=" * 70)

print(f"\nClient dataset shape: {datasets['client'].shape}")
print(f"Current columns: {list(datasets['client'].columns)}")
print(f"\nBirth number samples:")
print(datasets['client']['birth_number'].head(10).to_string())
print(f"\nBirth number data type: {datasets['client']['birth_number'].dtype}")
```

```python
from datetime import date

def parse_birth_number(birth_num):
    """
    Parse Czech birth number to extract birthday and gender.
    Format: YYMMDD (for males) or YY(MM+50)DD (for females)
    Example: 706213 -> 1970-12-13 (female, month 62 = 12 + 50)
             450204 -> 1945-02-04 (male)
    """
    birth_str = str(birth_num).zfill(6)
    year = int(birth_str[:2])
    month = int(birth_str[2:4])
    day = int(birth_str[4:6])
    
    # Check if female (month > 50)
    if month > 50:
        month -= 50
        gender = 'F'
    else:
        gender = 'M'
    
    # Determine century (00-30 = 2000s, 31-99 = 1900s)
    if year <= 30:
        year += 2000
    else:
        year += 1900
    
    birthday = date(year, month, day)
    
    return birthday, gender

print("\n" + "=" * 70)
print("Processing birth_number field...")
print("=" * 70)

# Apply the function to extract birthday and gender
datasets['client'][['birthday', 'gender']] = datasets['client']['birth_number'].apply(
    lambda x: pd.Series(parse_birth_number(x))
)

print(f"\nExtracted birthday and gender for {len(datasets['client'])} clients")

# Calculate age (as of 1998-12-31, the last date in the dataset)
reference_date = date(1998, 12, 31)
datasets['client']['age'] = datasets['client']['birthday'].apply(
    lambda bd: reference_date.year - bd.year - ((reference_date.month, reference_date.day) < (bd.month, bd.day))
)

print(f"Calculated age as of {reference_date}")
print("\n" + "=" * 70)
print("Birth number processing completed!")
print("=" * 70)
```

```python
print("\n" + "=" * 70)
print("CLIENT DATASET - After Processing")
print("=" * 70)

print(f"\nUpdated columns: {list(datasets['client'].columns)}")
print(f"\nSample data (first 10 rows):")
display(datasets['client'][['client_id', 'birth_number', 'birthday', 'gender', 'age', 'district_id']].head(10))

print("\n" + "=" * 70)
print("STATISTICS")
print("=" * 70)
print(f"\nGender distribution:")
print(datasets['client']['gender'].value_counts().to_string())

print(f"\n\nAge statistics:")
print(datasets['client']['age'].describe().to_string())

print(f"\n\nAge distribution:")
age_ranges = pd.cut(datasets['client']['age'], bins=[0, 20, 30, 40, 50, 60, 70, 150], 
                     labels=['0-20', '21-30', '31-40', '41-50', '51-60', '61-70', '70+'])
print(age_ranges.value_counts().sort_index().to_string())

print("\n" + "=" * 70)
```

**Birth Number Processing Summary:**
- Parsed birth_number field (YYMMDD format) to extract birthday and gender
- Gender identification: Month > 50 indicates female (month - 50 = actual month)
- Added 3 new columns to client dataset: `birthday`, `gender`, `age`
- Age calculated as of 1998-12-31 (end of dataset period)
- Client dataset now has complete demographic information

## Export Cleaned Datasets to CSV

Save each cleaned dataset to individual CSV files in the cleandata folder.

```python
# Create cleandata folder if it doesn't exist
output_folder = 'cleandata'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    print(f"Created folder: {output_folder}")
else:
    print(f"Using existing folder: {output_folder}")

print("\n" + "=" * 70)
print("EXPORTING CLEANED DATASETS TO CSV")
print("=" * 70)

# Save each dataset to CSV
export_summary = []
for dataset_name, df in datasets.items():
    output_file = os.path.join(output_folder, f"{dataset_name}.csv")
    
    try:
        # Save with semicolon delimiter to match original format
        df.to_csv(output_file, sep=';', index=False)
        
        file_size = os.path.getsize(output_file) / (1024 * 1024)  # Size in MB
        
        export_summary.append({
            'Dataset': dataset_name,
            'Rows': df.shape[0],
            'Columns': df.shape[1],
            'File Size (MB)': round(file_size, 2),
            'Status': 'Success'
        })
        
        print(f"Exported {dataset_name:15s} | {df.shape[0]:7,} rows × {df.shape[1]:2d} cols | {file_size:6.2f} MB")
        
    except Exception as e:
        export_summary.append({
            'Dataset': dataset_name,
            'Rows': df.shape[0],
            'Columns': df.shape[1],
            'File Size (MB)': 0,
            'Status': f'Error: {str(e)}'
        })
        print(f"Error exporting {dataset_name}: {str(e)}")

print("\n" + "=" * 70)
print("EXPORT SUMMARY")
print("=" * 70)
export_df = pd.DataFrame(export_summary)
print(export_df.to_string(index=False))
print("=" * 70)
print(f"\nTotal files exported: {len([s for s in export_summary if s['Status'] == 'Success'])}/{len(datasets)}")
print(f"Total size: {export_df['File Size (MB)'].sum():.2f} MB")
print(f"Output location: {os.path.abspath(output_folder)}")
print("=" * 70)
```

**Export Summary:**
- All cleaned datasets exported to `cleandata/` folder
- Files saved with semicolon (`;`) delimiter
- Includes all transformations:
  - Date columns converted to date format (account, loan, trans, card)
  - Client dataset enhanced with birthday, gender, and age columns
- Each dataset saved as individual CSV file
- Ready for further analysis or sharing
